```python
import os
os.environ['PYSPARK_SUBMIT_ARGS'] = '--packages io.delta:delta-spark_2.12:3.2.1,io.unitycatalog:unitycatalog-spark_2.12:0.2.0 --jars delta-spark_2.12-3.2.1.jar,unitycatalog-spark_2.12-0.2.0.jar pyspark-shell'
```

```python
from delta import *
import pyspark

builder = pyspark.sql.SparkSession.builder.appName("LocalUnityCatalogExperiment").master("local") \
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension") \
    .config("spark.sql.catalog.spark_catalog", "io.unitycatalog.spark.UCSingleCatalog") \
    .config("spark.sql.catalog.unity", "io.unitycatalog.spark.UCSingleCatalog") \
    .config("spark.sql.catalog.unity.uri", "http://127.0.0.1:3000") \
    .config("spark.sql.catalog.unity.token", "") \
    .config("spark.sql.defaultCatalog", "unity")

spark = configure_spark_with_delta_pip(builder).getOrCreate()

data = spark.range(0, 5)
data.show()
```
